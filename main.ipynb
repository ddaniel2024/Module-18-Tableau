```python
import pandas as pd
from pathlib import Path
from math import sin, cos, sqrt, atan2, radians
```

```python
jan_df1 = pd.read_csv("2023-citibike-tripdata/1_January/202301-citibike-tripdata_1.csv")
jan_df2 = pd.read_csv("2023-citibike-tripdata/1_January/202301-citibike-tripdata_2.csv")
jul_df1 = pd.read_csv("2023-citibike-tripdata/7_July/202307-citibike-tripdata_1.csv")
jul_df2 = pd.read_csv("2023-citibike-tripdata/7_July/202307-citibike-tripdata_2.csv")
jul_df3 = pd.read_csv("2023-citibike-tripdata/7_July/202307-citibike-tripdata_3.csv")
jul_df4 = pd.read_csv("2023-citibike-tripdata/7_July/202307-citibike-tripdata_4.csv")
```

```python
merged_df = pd.concat([jan_df1, jan_df2, jul_df1, jul_df2, jul_df3, jul_df4])
merged_df
```

```python
merged_df = merged_df.dropna()
merged_df.head()
```

```python
merged_df.dtypes
```

```python
merged_df["started_at"] = merged_df["started_at"].astype("datetime64[ns]")
merged_df["ended_at"] = merged_df["ended_at"].astype("datetime64[ns]")
```

```python
merged_df.dtypes
```

```python
merged_df = merged_df.sort_values(["started_at"])
merged_df.head()
```

```python
merged_df["duration"] = merged_df["ended_at"] - merged_df["started_at"]
merged_df["duration_minutes"] = round(merged_df["duration"].dt.seconds/60,1)
merged_df.head()
```

```python
merged_df.to_csv("output.csv", index=False)
```
